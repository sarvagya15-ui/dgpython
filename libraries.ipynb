```python
import math 
```

```python
print(math.sqrt(16))
print(math.pi)
print(math.gcd(48,18))
print(math.lcm(4,6))
print(math.factorial(5))
print(math.log(100,10))
print(math.sin(0))
```

```python
from random import random, randint,choice,shuffle

print("random number between 0 and 1:", random())
print("random integer between 1 and 10:",randint(1,10) )
print( "random choice from list:", choice([2,3,5,4,6,7,9,8,6,10]))
mylist=["John","jane","Jack","Julia"]
shuffle(mylist)
print("Shuffled list:",mylist )
```
