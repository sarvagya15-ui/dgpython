```python
a = [2,6,4,4,8,8,96,41,3,2,65,41,5,1]
for x in a :
    print(x)
```

```python
name  =  "David"
for c in name :
    print(c , end = "-")
    
# This is a comment
```

```python
for i in range (100,20,-6):
    print(i, end = " ")
```

```python
for i in range(5) :
    print(f"{i+1} sarvagya")
```

```python
for i in range(1,21) :
    if i % 2 != 0 :
        print(f"{i}", end = " ")
```

```python
for i in range(0,21,2):
    print(f"{i}" , end = " ")    
```

```python
for i in range(10):
    print( 2 * i)
```

```python
x = 0
while x < 10 :
    print(f"Value of x is {x}")
    x += 1 
```

```python
y = 0 
while y < 5 :
    print(f"sarvagya")
    y+=1
```

```python
z = 0
while z < 20 :
    if z % 2 != 0:
        print(f"{z}", end = " ")
    z += 1
```

```python
q = 0 
while q < 10 :
    print(2*q)
    q+=1
```

```python
j = "tat"
if j == j[::-1]:
    print("palindrome")
```

```python
n = 10  # Number of terms to print
a, b = 0, 1
i = 0

for i in range(n):
    print(a, end=' ')
    a, b = b, a + b
    i += 1
print() 
```
