```python
s1 = int(input("Enter first side of trinangle: "))
s2 = int(input("Enter second side of trinangle: "))
s3 = int(input("Enter third side of trinangle: "))

if s1 == s2 == s3:
    print("Equilateral triangle")
elif s1 == s2 and s2 != s3 or s2 == s3 and s1 != s2 or s1 == s3 and s2 != s3:
    print("Isosceles triangle")
else:
    print("Scalene triangle")
```
