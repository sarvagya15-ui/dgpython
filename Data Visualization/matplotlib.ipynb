```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('iris')
```

```python
#histogram - distribution of sepal length
plt.hist(df['sepal_length'], color='Orange', edgecolor='black',)
plt.title('Distribution of Sepal Length')
plt.xlabel('Sepal Length (cm)') 
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
#lineplot - petalength of first 50 samples
plt.plot(df['petal_length'][:50], marker='X', color='green',linestyle = "-.",markeredgecolor='black',markersize=8)
plt.title('Petal Length of First 50 Samples')
plt.xlabel('Sample Index')
plt.ylabel('Petal Length (cm)')
plt.show()
```

```python
mean_value = df.groupby('species').mean()
```

```python
mean_value.plot(kind='bar', color=['blue', 'orange', 'green','red'],figsize=(10,6))
plt.title('Average Count of Each Species')
plt.xlabel('Species')
plt.ylabel('Average Count') 
plt.xticks(rotation=0)
plt.show()
```

```python
import numpy as np
#creating  a custom plot
fig , ax = plt.subplots(figsize=(10,6))
 
#setting the color of figure and axes
fig.set_facecolor('blue')
ax.set_facecolor('lightgreen')

x = np.random.randint(0, 10, 20)
y = np.random.randint(0, 10, 20)

ax.plot(x,y,
        color='green',
        linewidth=2,
        marker='o',
        markerfacecolor='red',
        markeredgecolor='black',
        markersize=2,
        linestyle='--'
        )
ax.set_title('Custom Plot', fontsize=16, fontweight='bold')
ax.set_xlabel('X-axis', fontsize=12)
ax.set_ylabel('Y-axis', fontsize=12)
```
