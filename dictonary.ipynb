W A P to take name and phone number of 3 users and store them in dictionary

```python
phonebook = {}
for i in range(3):
    phonebook[input("name: ")] = input("phone number: ")
```

```python
phonebook
```

W A P  to find number in phonebook if not found printnot found

```python
x = input("name: ") 
if x in phonebook :
    print(phonebook.get(x))
else:
    print("name not found") 
```
