```python
#non paramiterized function
def greet():#function initialization
    print('hello world')
    print('welcome to python programming')
    print('let us learn python together and have fun')
    print('have a nice day')
greet()#function call
```

```python
#paramiterized function
def  sum(a,b):
    return a+b  
sum(9,6)
```

```python
import time


def average (a,b,c,d): #return type function
    return (a+b+c+d)/4
start = start_time = time.time()
avg = average(10,20,30,40)
print(avg)
end_time = time.time()
print(f"Execution time: {end_time - start_time} seconds")
```

```python
def country(name = "bharat"):
    print(f'I am from {name}')

country()
country('usa')  
country('canada')
```

```python
import time
square = lambda x: x**2
start = start_time = time.time()
print(square(5))
end = time.time()
print(f"Execution time: {end - start} seconds")
```

```python
lst = []
for i in range(1, 6):
    lst.append(int(input(f"Enter element {i}: ")))
print(lst)

lstsq = lambda lst:[x**2 for x in lst]


lstsq(lst)
```

```python
b = list(map(lambda x: x**2, lst))
print(b)
```
